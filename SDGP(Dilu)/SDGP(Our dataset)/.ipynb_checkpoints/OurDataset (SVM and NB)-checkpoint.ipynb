### Feature Selection / Corelation

```python
X = df.drop("Marks",axis = 1)
y = df["Marks"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X , y , test_size = 0.3, random_state = 0)

X_train.shape, X_test.shape
```

```python
X_train.corr()
```

```python
plt.figure(figsize = (50,100))
cor = X_train.corr()
sns.heatmap(cor, annot = True, cmap = plt.cm.CMRmap_r)
```

```python
# function to check highly correlated features

def correlation(dataset, threshold):
    col_corr = set() #Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range (i):
            if abs(corr_matrix.iloc[i,j]) > threshold : 
                colname = corr_matrix.columns[i]
                col_corr.add(colname) #getting the name of column
    return col_corr
```

```python
corr_features = correlation(X_train , 0.6)
len(set(corr_features))
#No highly co-related feeatures were found
```

### SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
```

```python
from collections import Counter
print("Before SMOTE:", Counter(y_train))

print("After SMOTE: ", Counter(y_train_smote))
```

### K-fold cross validation

```python
from sklearn.model_selection import StratifiedKFold
#10 folds
kf = StratifiedKFold(n_splits= 10)
kf
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test,y_test)
```

```python
get_score(svm.SVC(decision_function_shape='ovo'), X_train, X_test, y_train, y_test)
```

```python
scores_svm = []
for train_index, test_index in kf.split(X,y):
    X_train, X_test, y_train, y_test = X.iloc[train_index], X.iloc[test_index], y.iloc[train_index], y.loc[test_index]
    
    scores_svm.append(get_score(svm.SVC(decision_function_shape='ovo'), X_train, X_test, y_train, y_test))
```
